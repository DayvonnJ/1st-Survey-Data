## SEO Experience Graphs

```python
# Load your survey data from the specified CSV file
df = pd.read_csv('Scholar_Survey_Responses_Spring_2015.csv')
# Define the statements and their corresponding response categories
statements = [
    "I enjoy coming to SEO most of the time.",
    "I try to do my best in my SEO courses.",
    "My instructorsâ€™ expectations make me want to do my best.",
    "What I learn in class helps me outside of SEO.",
    "I take pride in my SEO work.",
    "I am getting a high-quality education through SEO.",
    "My classmates at SEO are supportive of me.",
    "I am comfortable seeking assistance from my classmates at SEO.",
    "I feel connected to my peers at SEO."
]

response_categories_1 = ["Strongly Disagree", "Disagree", "Agree", "Strongly Agree"]

# Initialize lists to store the counts for each response category
response_counts = {category: [] for category in response_categories_1}

# Iterate through the statements
for statement in statements:
    # Get the value counts for the current statement
    cnt_srs_1 = df[statement].value_counts()
    
    # Append the counts for each response category to the respective list
    for category in response_categories_1:
        response_counts[category].append(cnt_srs_1.get(category, 0))

# Create bar chart traces for each response category
traces = []
for category in response_categories_1:
    trace = go.Bar(
        y=statements,
        x=response_counts[category],
        orientation='h',
        name=category
    )
    traces.append(trace)

# Create the layout
layout = go.Layout(
    title='SEO Experience Review',
    barmode='stack',
    width=1000,
    height=600
)

# Create the figure
fig = go.Figure(data=traces, layout=layout)

# Show the plot
fig.show()
# Save the figure as an HTML file
fig.write_html("seo_experience.html")
#Saving Image as JPEG
fig.write_image("seo_experience.jpeg")
```
